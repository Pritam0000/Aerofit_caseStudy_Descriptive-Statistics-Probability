**Aerofit**

Aerofit stands out as a prominent name in the realm of fitness gear. Aerofit offers a diverse selection of products, encompassing machinery like treadmills, stationary bikes, gym apparatus, and accessories designed to meet the requirements of various demographics.

**Business Challenge**

The market analysis department at Aerofit aims to discern the distinctive attributes of the target demographic for each treadmill variant offered by the company, aiming to enhance recommendations for new customers. The team opts to explore potential divergences in customer characteristics across the product range.

**Approach**

Utilize descriptive analytics to formulate a customer profile for every Aerofit treadmill product through the creation of suitable tables and charts. Generate two-way contingency tables for each AeroFit treadmill product, calculating all conditional and marginal probabilities, and elucidate their implications on business operations.

**Features**

The dataset encompasses 180 data points featuring 9 attributes.


1. Product Acquired: KP281, KP481, or KP781
2. Age: Quantified in years
3. Gender: Male or Female
4. Educational Attainment: Number of years of education completed
5. Marital Status: Single or in a partnership
6. Usage Frequency: Average number of times per week the customer plans to utilize the treadmill
7. Income: Annual earnings in dollars
8. Physical Fitness: Self-assessed fitness level rated on a scale from 1 to 5, where 1 signifies poor condition and 5 denotes excellent condition
9. Distance: Expected average weekly distance the customer aims to walk or run
